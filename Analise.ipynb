```python
def count(df, column):
    df_novo = pd.DataFrame(df[column].value_counts())
    df_novo_percent = pd.DataFrame(round(df_novo * 100 / df_novo.sum()))
    df_comp = df_novo_percent.join(df_novo, lsuffix='%')
    df_comp.loc['Total'] = df_comp.sum()

    return df_comp
```

```python
def df_graph(col1, col2):
    df_novo = pd.DataFrame(
        round(
            dados.groupby([col1,
                                 col2])[col1].count().unstack(col2).fillna(0)))

    df_novo_percent = pd.DataFrame(round(df_novo * 100 / df_novo.sum(),1))

    df_comp = df_novo_percent.join(df_novo, lsuffix='%')
    df_comp.loc['Total'] = df_comp.sum()

    return df_comp
```
