##### Funções e tratamento

```python
#verificar os nulos
def nulos(df):
    nan = df.isna().sum()
    return print(nan)
```

```python
#Quantidade de dados duplicados
def duplicados(df):
    duplos = df.duplicated(keep = False).value_counts()
    return duplos
```

```python
def count(df,column):
    return df[column].value_counts()
```

```python
def percent(df,column,total):
    return round(df[column].value_counts()*100/total)
```

```python
nulos(dados)
```

```python
dados.drop(['progress','motvimigr'], axis=1, inplace=True)
```

```python
for (row, col) in dados.iterrows():
    if col.raca == 'Preta':
        dados['raca'].replace(to_replace=col.raca, value='Negra', inplace=True)
    if col.raca == 'Parda':
        dados['raca'].replace(to_replace=col.raca, value='Negra', inplace=True)
```

```python
dados['ondmora'] = dados['ondmora'].str.strip()
```

```python
dados['ondmora'] = dados['ondmora'].str.replace('Criado no suburbio - Pavuna', 'Pavuna')
dados['ondmora'] = dados['ondmora'].str.replace('ramos', 'Ramos')
```

```python
dados['idade'].fillna(-1,inplace=True)
dados['orientsex'].fillna(-1,inplace=True)
dados.fillna('Sem informação',inplace=True)
```

```python
def agrupando_gen(df,column):
    df_novo = pd.DataFrame(round(df.groupby([column, 'idgen'])[column].count().unstack('idgen').fillna(0)))
    df_novo = df_novo.astype(int)
    return df_novo
```

```python
def agrupando_raca(df,column):
    df_novo = pd.DataFrame(round(df.groupby([column, 'raca'])[column].count().unstack('raca').fillna(0)))
    df_novo = df_novo.astype(int)
    df_novo['Total'] = df_novo['Branca'] + df_novo['Indígena'] + df_novo['Negra']
    
    return df_novo
```
